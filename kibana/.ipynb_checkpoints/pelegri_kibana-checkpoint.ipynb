## Working on all dimensions : je voulais construire un dataset en finance : pour avoir la consistence necéssaire pour constuire le modèle en étoile j'ai fait une database très "basique" et les calcules à la main pour enrichir tout ca

```python
google = pd.read_csv("GOOG.csv") 
amazon = pd.read_csv("AMZN.csv") 
apple = pd.read_csv("AAPL.csv") 
facebook = pd.read_csv("FB.csv") 
```

```python
google.head()
```

```python
n = len(google['Open'])
X = [ i for i in range(n)]
google['id_google'] = X
google.set_index('id_google')
google.head()
```

```python
n = len(google['Open'])
print(n)
```

```python
n = len(facebook['Open'])
print(n)
```

```python
n = len(amazon['Open'])
print(n)
```

```python
n = len(apple['Open'])
print(n)
```

```python
apple['id_apple'] = X
apple.set_index('id_apple')

amazon['id_amazon'] = X
amazon.set_index('id_amazon')

facebook['id_facebook'] = X
facebook.set_index('id_facebook')

facebook.tail()
```

```python
google['variation'] = ( (google['High'] - google['Low']) /  google['High'] ) * 100
amazon['variation'] = ( ( amazon['High'] - amazon['Low'] ) / amazon['High'] ) * 100
apple['variation'] = ( ( apple['High'] - apple['Low'] ) / apple['High'] ) * 100
facebook['variation'] = ( ( facebook['High'] - facebook['Low'] ) / facebook['High'] ) * 100

google.head()
```

```python
google["weekday"] = pd.to_datetime(google.Date).dt.dayofweek
apple["weekday"] = pd.to_datetime(apple.Date).dt.dayofweek
facebook["weekday"] = pd.to_datetime(facebook.Date).dt.dayofweek
amazon["weekday"] = pd.to_datetime(amazon.Date).dt.dayofweek

google.head()
```

```python
facebook.to_csv('facebook.csv')
google.to_csv('google.csv')
amazon.to_csv('amazon.csv')
apple.to_csv('apple.csv')

print('all dimensions prepared for kibana')
```

## Looking att portfolio now : the center tabke

```python
portfolio = pd.read_csv("portfolio.csv") 
portfolio['date'] = google['Date']

portfolio['id_google'] = google['id_google']
portfolio['id_facebook'] = facebook['id_facebook']
portfolio['id_amazon'] = amazon['id_amazon']
portfolio['id_apple'] = apple['id_apple']

portfolio.head()
```

```python
portfolio['mean_variation'] = google['variation'] + facebook['variation'] + amazon['variation'] + apple['variation']
portfolio['mean_variation'] = portfolio['mean_variation'] / 4

portfolio.head()
```

```python
portfolio['mean_volume'] = np.nan

portfolio['mean_volume'] = ( google['Volume'] + facebook['Volume'] + amazon['Volume'] + apple['Volume'] ) / 4 

portfolio.head()
```

```python
portfolio['mean_variation_friday'] = np.nan

portfolio['mean_variation_friday'] = ( google.loc[google['weekday'] == 4]['variation'] + facebook.loc[facebook['weekday'] == 4]['variation'] + amazon.loc[amazon['weekday'] == 4]['variation'] + apple.loc[apple['weekday'] == 4]['variation'] )
portfolio.head()
```

```python
for i in range(len(portfolio['mean_variation_friday']) - 1):
    if type(portfolio['mean_variation_friday'][i]) == 'float':
        portfolio['mean_variation_friday'][i] = portfolio['mean_variation_friday'][i] / 4
        
portfolio.head()
```

```python
portfolio['mean_variation_friday'] = portfolio['mean_variation_friday'].replace(np.nan, 0)
portfolio.head()
```

### Portefolio 1 : 2 500 on each stock : 
- 4.6 action google

```python
for i in range(len(portfolio['portfolio_1']) - 1):
    portfolio['portfolio_1'][i] = 4.6 * google['Open'][i] + 30 * facebook['Open'][i] + 6.3 * amazon['Open'][i] + 20 * apple['Open'][i]
    
portfolio.tail()
```
